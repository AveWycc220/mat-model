```python
import numpy as np
import matplotlib.pyplot as plt
import math

%matplotlib inline
```

```python
t=np.arange(20)
a0=1
a1=2
a2=3
a3=4

y=a0+a1*t+a2*t**2+a3*t**3

plt.figure(figsize=(10, 5))
plt.scatter(t, y)
plt.show()
```

```python
def MNK_line(x, y):
    n=len(x)
    
    x2=x**2
    
    sumXY=0
    for i in range(len(x)):
        sumXY=sumXY+x[i]*y[i]
    
    k=(n*sumXY-sum(y)*sum(x))/(n*sum(x2)-sum(x)**2)
    
    b=(sum(y)/n)-k*(sum(x)/n)

    return k, b

def MNK_parabola(x, y):
    N=len(x)
    
    x2=x**2
    x3=x**3
    x4=x**4
    
    S1=sum(x)
    S2=sum(x2)
    S3=sum(x3)
    S4=sum(x4)
    
    Sy0=sum(y)
    
    Sy1=0
    for i in range(len(x)):
        Sy1=Sy1+x[i]*y[i]
        
    Sy2=0
    for i in range(len(x)):
        Sy2=Sy2+x2[i]*y[i]
        
    A = np.array([[N, S1, S2], [S1, S2, S3], [S2, S3, S4]])
    A_inv = np.linalg.inv(A)
    B=np.array([[Sy0], [Sy1], [Sy2]])
    
    C = A_inv.dot(B)

    return C[0], C[1], C[2]

def MNK_tr(x, y):
    N=len(x)
    
    x2=x**2
    x3=x**3
    x4=x**4
    x5=x**5
    x6=x**6
    
    S1=sum(x)
    S2=sum(x2)
    S3=sum(x3)
    S4=sum(x4)
    S5=sum(x5)
    S6=sum(x6)
    
    Sy0=sum(y)
    
    Sy1=0
    for i in range(len(x)):
        Sy1=Sy1+x[i]*y[i]
        
    Sy2=0
    for i in range(len(x)):
        Sy2=Sy2+x2[i]*y[i]
        
    Sy3=0
    for i in range(len(x)):
        Sy3=Sy3+x3[i]*y[i]
        
    A = np.array([[N, S1, S2, S3], [S1, S2, S3, S4], [S2, S3, S4, S5], [S3, S4, S5, S6]])
    A_inv = np.linalg.inv(A)
    B=np.array([[Sy0], [Sy1], [Sy2], [Sy3]])
    
    C = A_inv.dot(B)

    return C[0], C[1], C[2], C[3]
```

```python
k, b=MNK_line(t, y)
print("Уравнение прямой:", b, '+', k, '* x')

b0, b1, b2=MNK_parabola(t, y)
print('Уравнение параболы:', b0[0], '+', b1[0], '* x+', b2[0], '* x^2')

c0, c1, c2, c3=MNK_tr(t, y)
print('Уравнение параболы:', c0[0], '+', c1[0], '* x+', c2[0], '* x^2 +', c3[0], '* x^3')

x = np.linspace(0, len(t), 100)

plt.figure(figsize=(10, 5))
plt.scatter(t, y)
plt.plot(x, b+k*x, 'b', label='Прямая')
plt.plot(x, b0+b1*x+b2*x*x, 'g', label='Парабола')
plt.plot(x, c0+c1*x+c2*x*x+c3*x*x*x, 'r', label='Полином 3-ей степени')
plt.legend()
plt.show()
```

```python
t=np.arange(20)
a0=1
a1=2
a2=3
a3=4

y=[]
for i in range(len(t)):
    y.append(a0+a1*math.sin(t[i])+a2*math.cos(t[i])+a3*t[i])

plt.figure(figsize=(10, 5))
plt.scatter(t, y)
plt.show()
```

```python
k, b=MNK_line(t, y)
print("Уравнение прямой:", b, '+', k, '* x')

b0, b1, b2=MNK_parabola(t, y)
print('Уравнение параболы:', b0[0], '+', b1[0], '* x +', b2[0], '* x^2')

c0, c1, c2, c3=MNK_tr(t, y)
print('Уравнение параболы:', c0[0], '+', c1[0], '* x+', c2[0], '* x^2 +', c3[0], '* x^3')

x = np.linspace(0, len(t), 100)

plt.figure(figsize=(10, 5))
plt.scatter(t, y)
plt.plot(x, b+k*x, 'b', label='Прямая')
plt.plot(x, b0+b1*x+b2*x*x, 'g', label='Парабола')
plt.plot(x, c0+c1*x+c2*x*x+c3*x*x*x, 'r', label='Полином 3-ей степени')
plt.legend()
plt.show()
```
